# 2.1.1 변수란 무엇일까요?

변수는 바구니이다.

`변수이름 = 데이터`

```python
a = 5    # a라는 바구니에 값 5를 넣는다. [Enter]
a         # a 안에 있는 값을 확인한다.     [▶Run 또는 Shift + Enter]
```

```python
b = 3    # b라는 바구니에 값 3을 넣는다. [Enter]
b         # b 안에 있는 값을 확인한다.     [▶Run 또는 Shift + Enter]
```

```python
a + b    # a 바구니에 있는 값과 b 바구니에 있는 값을 더한다.   [▶Run 또는 Shift + Enter]
```

```python
c = '가나다'
c
```

```python
radio_freq = 107.9
radio_freq
```
