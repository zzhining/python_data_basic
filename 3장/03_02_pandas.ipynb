```python
def getDate(data):
    data = str(data)
    year = data[0:4]
    month = data[4:6]
    day = data[6:8]
    return year+'-'+month+'-'+day
```

### merge

- inner join
- left join
- right join
- full outer join

    []  
      .

```python
df1 = pd.DataFrame({'': ['A', 'B', 'D'],
                    '':['', '', '']
                   })
df1
```

```python
df2 = pd.DataFrame({'': ['A', 'B', 'C'],
                    '':[180, 160, 150],
                   })
df2
```

```python
pd.merge(df1, df2)
```

----

image-20210606232521550.png

```python
pd.merge(df1, df2, how='left')
```

----

image-20210606232543046.png

```python
pd.merge(df1, df2, how='right')
```

----

image-20210606232557458.png

```python
pd.merge(df1, df2, how='outer')
```
