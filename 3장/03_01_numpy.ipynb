sort: 

```python
arr = np.random.randn(5)
print(' :', arr)
arr.sort()
print(' :', arr)
```

```python
arr = np.random.randn(9).reshape(3, 3)
print(' :\n', arr)
arr.sort(axis = 1)
print(' :\n', arr)
```

:   + 

```python
arr = np.random.randn(100)
print(arr)
arr.sort()

quantile_1 = arr[int(0.25 * len(arr))]
quantile_4 = arr[int(0.75 * len(arr))]

print('\n\n25%:', quantile_1, '75%:', quantile_4)
```
