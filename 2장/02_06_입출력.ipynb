# 2.6.3 파일 입출력

```python
f = open("abc.txt", "w") # 쓰기모드로 파일 열기
f.write("A B C D E F G ")
f.close()
```

```python
f = open("abc.txt", "r")
print(f.read())
f.close()
```

```python
f = open("abc.txt", "w")
f.write("a b c d e f g ")
f.close()
```

```python
f = open("abc.txt", "r")
print(f.read())
f.close()
```

```python
# 추가(append)모드로 파일 열기
f = open("abc.txt", "a")
f.write("H I J K L M N O P Q R S T U V W X Y Z")
f.close()
```

```python
# 읽기모드로 파일 열기
f = open("abc.txt", 'r')
lines = f.readlines()
for line in lines:
    print(line)
f.close()
```

```python
with open("일기.txt", "w") as f:
    f.write("2020년 3월 12일 금요일\n")

with open("일기.txt", "a") as f:
    f.write("날씨 맑음")

with open("일기.txt", "r") as f:
    print(f.read())
```

```python
import os

fileName = "abc.txt"
if os.path.exists(fileName):
    os.remove(fileName)
    print("{}파일을 삭제하였습니다.".format(fileName))
else:
    print("{}파일이 존재하지 않습니다.".format(fileName))
```
